```python
class PlayerInformation():
    def __init__(self):
        self.name = input("Enter player Name: ")
        
        self.age = int(input("Enter player Age: "))
        if self.age < 18:
            print("You are not allowed to enter the casino.")
            self.eligible = False
        else:
            self.eligible = True
            
        if self.eligible == True:
            self.balance = int(input("\nEnter your current bank balance: "))
        
            while True:
                self.bet = int(input("Enter the amount you want to bet: "))
                if self.bet>self.balance:
                    print("Not enough credits, please try with a lower amount.")
                    continue
                else:
                    self.balance-=self.bet
                    break
        
```

```python
class Cards():
    def __init__(self):
        self.cards_base = ["Ace", 2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King"]
        self.suite_base = ["Spade","Hearts","Diamonds","Clubs"]

        self.cards = []
        
        for base in self.suite_base:
            for card in self.cards_base:
                card_value = str(card) + " of " + str(base)
                self.cards.append(card_value)
        
    def shuffle_cards(self):
        import random
        random.shuffle(self.cards)
        
    def draw_card(self):
        card = self.cards.pop()
        if (card[0] == 1 or card[0] == "J" or card[0] == "Q" or card[0] == "K"):
            value = 10
        else:
            value = card[0]
        
        return (card,value)
    
    def display_cards(self,dealer_cards,player_cards):
        self.dealer_cards = dealer_cards
        self.player_cards = player_cards
        
        self.dealer_rounds = len(self.dealer_cards)
        self.player_rounds = len(self.player_cards)
        
        print("______________________________________________")
        print("\nDealer Cards:")
        for i in range(0,self.dealer_rounds):
            print(f"{self.dealer_cards[i][0]}")
            
        print("\nPlayer Cards:")
        for i in range(0,self.player_rounds):
            print(f"{self.player_cards[i][0]}")
        print("______________________________________________")
```

```python
class Game():
    def __init__(self):
        self.player = PlayerInformation()
        self.cards = Cards()
        self.cards.shuffle_cards()
        
        self.dealer_cards = []
        self.player_cards = []
        self.game_won = 0
        self.game_lost = 0
        
        self.dealer_cards.append(self.cards.draw_card())
        self.dealer_cards.append(self.cards.draw_card())
        
        self.player_cards.append(self.cards.draw_card())
        self.player_cards.append(self.cards.draw_card())
        
        if self.player.eligible == True:
            print(f"\nDealer cards:\n {self.dealer_cards[0][0]} : HIDDEN")
            print(f"Player cards:\n {self.player_cards[0][0]} : {self.player_cards[1][0]}")
        
    def win_game(self,turn):
        self.turn = turn
        self.player_rounds = len(self.player_cards)
        self.player_score = 0
        for i in range(0,self.player_rounds):
            value = self.player_cards[i][1]
            if value == "A":
                if self.player_score > 11:
                    value = 1
                else:
                    value = 10
            self.player_score+=int(value)
            
        self.dealer_rounds = len(self.dealer_cards)
        self.dealer_score = 0
        for i in range(0,self.dealer_rounds):
            value = self.dealer_cards[i][1]
            if value == "A":
                if self.dealer_score > 11:
                    value = 1
                else:
                    value = 10
            self.dealer_score+=int(value)
            
        if self.turn == "Player":
            if self.player_score > 21:
                print("\nBust! Dealer Wins.")
                self.game_lost+=1
                print(f"Player balance: {self.player.balance}")
                return False
            else:
                return True

        else:
            if self.dealer_score > 21:
                print("\nBust! Player Wins.")
                self.game_won+=1
                self.player.balance+=(self.player.bet*2)
                print(f"Player balance: {self.player.balance}")
                return False
            elif self.dealer_score > self.player_score:
                print("\nDealer beats the player.")
                self.game_lost+=1
                print(f"Player balance: {self.player.balance}")
                return False
            else:
                return True
    
    def play_game(self):
        while True:
            play = input("Do you want to Hit or Stay? ")
            if play.lower() == "hit":
                self.player_cards.append(self.cards.draw_card())
                self.cards.display_cards(self.dealer_cards,self.player_cards)
                break
            elif play.lower() == "stay":
                break
            else:
                print("\nInvalid input, please try again.")
                       
        if self.win_game("Player"):
            while self.win_game("Dealer"):
                self.dealer_cards.append(self.cards.draw_card())
                self.cards.display_cards(self.dealer_cards,self.player_cards)
        else:
            pass   
```
